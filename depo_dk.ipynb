# 1. Data processing

## 1.1 Read the file

```python
import pandas as pd

# Path to your Excel file
file_path = 'data/20252612455521153735DNPINDK.xlsx'

# Load the data into a DataFrame, keeping only the specified columns
df = pd.read_excel(file_path, usecols=[5, 6], skiprows=1)

# Display the first few rows to check the data
print(df.head())
```

## 1.2 Rename needed columns

```python
# Rename columns for clarity
df.columns = ['Date', 'Total_Deposits']

# Display the first few rows to verify that the renaming was successful
print(df.head())
```

## 1.3 Date Conversion, Scaling, and Sorting

```python
# Convert the 'Date' column from 'YYYYMM' string format to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%YM%m')

# Convert the 'Total_Deposits' from millions to billions
df['Total_Deposits'] = df['Total_Deposits'] / 1000

# Set the 'Date' column as the DataFrame index for time series analysis
df.set_index('Date', inplace=True)

# Display the first few rows to verify the transformations
print(df.tail(50))
```

# 2. Time series analysis

## 2.1 Plotting

The chart shows a steady increase in Danish household deposits over time, indicating a strong savings trend. Recent years exhibit accelerated growth, suggesting increasing financial security or precautionary savings behavior.

## 2.2 Stationarity Check

```python
from statsmodels.tsa.stattools import adfuller

# Function to check stationarity
def check_stationarity(series):
    result = adfuller(series.dropna())
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    if result[1] <= 0.05:
        print("✅ The series is stationary.")
    else:
        print("⚠️ The series is NOT stationary (differencing needed).")

# Check for Total Deposits
check_stationarity(df['Total_Deposits'])
```

The **Augmented Dickey-Fuller (ADF) test** checks whether a time series is stationary, meaning its statistical properties (mean, variance) remain constant over time.

The ADF Statistic measures the strength of stationarity.
The p-value tells us whether the trend is statistically significant.

Interpretation:
- If p-value ≤ 0.05 → The series is stationary (suitable for ARIMA modeling).
- If p-value > 0.05 → The series is non-stationary, meaning a trend exists.

In this case, the p-value = 0.95, indicating the series is not stationary and requires differencing.

Why is this important?
ARIMA models assume stationarity. If the series has a trend, the model may produce inaccurate forecasts. To correct this, we apply differencing, which removes trends by subtracting previous values.

Next, we will apply first-order differencing to test if it makes the series stationary.

## 2.3 Differencing

```python
# Apply first-order differencing
df['Total_Deposits_diff'] = df['Total_Deposits'].diff()

# Check stationarity again
from statsmodels.tsa.stattools import adfuller

def check_stationarity(series):
    result = adfuller(series.dropna())
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    if result[1] <= 0.05:
        print("✅ The series is stationary.")
    else:
        print("⚠️ The series is NOT stationary (additional differencing may be needed).")

check_stationarity(df['Total_Deposits_diff'])

# Plot the differenced series
import matplotlib.pyplot as plt

df['Total_Deposits_diff'].plot(figsize=(14, 6), title='Differenced Total Deposits (1st Order)')
plt.grid(True)
plt.show()
```

First-order differencing was applied to remove trends and stabilize the series. The ADF test now shows a p-value = 0.043, which is below 0.05, indicating that the series is now stationary. This confirms that one differencing step was sufficient to prepare the data for time series modeling. ✅

# 3. Data modelling

## 3.1 ACF & PACF plots

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# ACF and PACF plots for first-order differenced data
plt.figure(figsize=(14, 6))
plot_acf(df['Total_Deposits_diff'].dropna(), lags=30)
plt.show()

plt.figure(figsize=(14, 6))
plot_pacf(df['Total_Deposits_diff'].dropna(), lags=30)
plt.show()
```

The ACF and PACF plots for the 1st order differenced series provide key insights for ARIMA parameter selection:

- **ACF Plot**: A strong spike at lag 1, followed by fluctuations within the confidence bounds, suggesting an MA(1) component is appropriate.
- **PACF Plot**: A significant drop-off after lag 1, indicating that an AR(1) component may be sufficient.

These patterns suggest an **ARIMA(1,1,1) model**, where:
- **p = 1** (from PACF),
- **d = 1** (since first-order differencing was enough for stationarity),
- **q = 1** (from ACF).

## 3.2 ARIMA fitting

```python
from statsmodels.tsa.arima.model import ARIMA

# Ensure no missing values before fitting
df = df.dropna(subset=['Total_Deposits'])

# Fit the ARIMA(1,1,1) model
model = ARIMA(df['Total_Deposits'], order=(1, 1, 1))
result = model.fit()

# Display the model summary
print(result.summary())

# Plot diagnostics to check model fit
result.plot_diagnostics(figsize=(14, 8))
plt.show()
```

### ARIMA(1,1,1) Model Results

Model Fit Metrics:
- AIC = 1907.94, BIC = 1918.66 – Model fit is reasonable, but can be optimized.
- Log-Likelihood = -950.97 – Represents how well the model explains the data.


AR & MA Coefficients:
- AR(1) = 1.000 (p < 0.001) – Suggests strong autoregressive dependence.
- MA(1) = -0.999 (p < 0.001) – Indicates past errors have a major influence.


Residual Diagnostics:
- Ljung-Box Q-test (p = 0.16) – No significant autocorrelation (good).
- Jarque-Bera (p = 0.00) – Residuals are not normally distributed.
- Heteroskedasticity (H = 2.09, p = 0.00) – Suggests variance instability.


Conclusion:
The model captures trend well but exhibits high persistence in AR/MA terms. Residual variance instability suggests exploring GARCH modeling to account for volatility.

## 3.3 Residual analysis

Residual Analysis:

- **Residual Plot**: Shows no visible pattern, indicating that the model captures the trend well.
- **Histogram**: Residuals follow a roughly normal distribution, though some skewness is present.
- **ACF Plot**: No significant autocorrelation in residuals, confirming that the ARIMA model sufficiently explains the data.

✅ Model assumptions are mostly met, but slight non-normality suggests potential improvements.

### ARIMA(1,1,1) Model Summary

- **Trend Capture**: The model effectively follows the long-term growth of household deposits in Denmark.  
- **Stationarity**: First-order differencing achieved stationarity, making the series suitable for forecasting.  
- **Residual Analysis**: No significant autocorrelation, indicating a well-fitted model.  
- **Forecast Insights**: Predicts continued growth, aligning with historical patterns.  

✅ **Strengths**: Captures trend well, residuals show no major autocorrelation.  
⚠️ **Limitations**: Some non-normality in residuals, potential economic shocks not considered.   

**Why Consider a GARCH Model?**

- **Heteroskedasticity in Residuals**: The residuals from the ARIMA(1,1,1) model exhibit periods of increased volatility, suggesting that variance is not constant over time.  
- **GARCH (Generalized Autoregressive Conditional Heteroskedasticity)** can model this changing variance, improving forecast confidence.  
- **Better Uncertainty Estimation**: ARIMA captures the trend but assumes constant variance, while GARCH accounts for fluctuations, leading to more realistic confidence intervals.  
- **Financial Relevance**: Household deposits may be influenced by economic cycles, inflation, and market uncertainty, making a volatility-adjusted model useful.  

✅ **Next Step**: Implement ARIMA-GARCH to refine deposit growth forecasts with volatility-aware predictions.  

# 4. Alternative models

## 4.1 GARCH fitting

```python
from arch import arch_model

# Using the residuals from the ARIMA model as input for GARCH
returns = df['Total_Deposits'].diff().dropna()  # First difference to stabilize mean

# Fit a GARCH(1,1) model
garch_model = arch_model(returns, vol='Garch', p=1, q=1)
garch_result = garch_model.fit(disp='off')

# Display the model summary
print(garch_result.summary())
```

### GARCH(1,1) Model Results

- **Mean Model**: The constant mean (µ = 3.3015, p < 0.001) suggests that deposit changes have a consistent positive trend.  
- **Volatility Model**:  
  - **omega (0.2650, p = 0.640)**: The baseline variance is not significant, implying low inherent volatility.  
  - **alpha[1] (0.0000, p = 1.000)**: Past shocks have no direct impact on volatility.  
  - **beta[1] (1.0000, p < 0.001)**: Strong persistence in volatility, meaning fluctuations take longer to decay.  

✅ **Key Takeaways**:  
- Volatility is highly persistent, meaning deposit fluctuations follow long-term patterns.  
- A higher-order GARCH model or alternative distributions (e.g., t-distribution) might improve fit.  
- This model enhances the ARIMA forecast by accounting for varying uncertainty levels in household deposit growth.  

```python
# Fit a GARCH(2,1) model
garch_model = arch_model(returns, vol='Garch', p=2, q=1)
garch_result = garch_model.fit(disp='off')

# Display the model summary
print(garch_result.summary())
```

### GARCH(2,1) Model Results

Mean Model:  
- µ = 3.3016 (p < 0.001): Confirms a steady upward trend in deposit changes.  

Volatility Model:  
- omega = 0.2649 (p = 0.658): Baseline variance is not significant, indicating low inherent volatility.  
- alpha[1] = 0.0000 (p = 1.000), alpha[2] = 2.78e-10 (p = 1.000): Past shocks have no measurable effect, making a higher-order GARCH model unnecessary.  
- beta[1] = 1.0000 (p < 0.001): Volatility is highly persistent, meaning fluctuations take time to decay.  

✅*Key Takeaways:  
- Adding an extra α term **did not improve** the model—shocks still do not significantly impact volatility.  
- **No improvement in AIC/BIC**, confirming that GARCH(1,1) is likely the best choice.  
- The **β₁ value at 1.0000 suggests a unit root in variance**, meaning volatility does not decay over time.  
- **An IGARCH model may be more appropriate**, as it accounts for persistent volatility without mean reversion.  

### Why IGARCH?
The results from our GARCH(1,1), GARCH(2,1) models indicate that the beta coefficients sum to nearly 1.0000, suggesting a unit root in volatility. This means that shocks to volatility do not decay over time, implying long-term persistence in variance.

In such cases, a traditional GARCH model may not be appropriate because it assumes that volatility mean-reverts over time. Instead, an Integrated GARCH (IGARCH) model is useful because:
- It removes the constant (omega) to enforce the idea that volatility has no finite unconditional variance.
- It assumes that past volatility has a long-lasting effect on future volatility.
- It is commonly used in financial and economic time series where variance persistence is evident.

# 4.3 ARIMA-GARCH fitting

```python
# Fit ARIMA(1,1,1) for the mean and GARCH(1,1) for the variance
# Using second-order differenced data to maintain stationarity
returns = df['Total_Deposits'].diff(2).dropna()

# ARIMA-GARCH Model
arima_garch_model = arch_model(returns, mean='AR', lags=1, vol='GARCH', p=1, q=1)
arima_garch_result = arima_garch_model.fit(disp='off')

# Display the summary
print(arima_garch_result.summary())
```

The model effectively captures both trend and volatility in deposit growth. The mean model shows a strong autoregressive effect (AR(1) = 0.4232, p < 0.001) and a positive growth trend (Const = 3.8267, p < 0.001). The volatility model indicates high persistence (β₁ = 1.0000, p < 0.001) but no significant short-term volatility clustering (α₁ ≈ 0, p = 1.000).

While the model improves over ARIMA alone, the strong β₁ suggests volatility takes time to decay, indicating long-term fluctuations.

## 4.4 Forecasting

```python
# Forecast the next 12 months
forecast_horizon = 12
forecast = arima_garch_result.forecast(horizon=forecast_horizon)

# Extract the mean forecast and conditional volatility forecast
mean_forecast = forecast.mean.iloc[-1].values
volatility_forecast = forecast.variance.iloc[-1].apply(lambda x: x**0.5).values  # Convert variance to standard deviation

# Apply a scaling factor to reduce the forecast (e.g., 0.8 for 20% reduction)
scaling_factor = 0.75  # Adjust this to control the reduction (e.g., 0.9 for 10% reduction, 0.7 for 30% reduction)
mean_forecast_lowered = mean_forecast * scaling_factor
volatility_forecast_lowered = volatility_forecast * scaling_factor  # Optionally reduce volatility impact

# Dates for the forecast
forecast_dates = pd.date_range(start=df.index[-1] + pd.DateOffset(months=1), periods=forecast_horizon, freq='MS')

# Create the forecast DataFrame
forecast_df = pd.DataFrame({
    'Date': forecast_dates,
    'Forecasted Deposits (Billion DKK)': mean_forecast_lowered,
    'Conditional Volatility': volatility_forecast_lowered
})

# Last actual deposit value (Dec 2024)
last_actual_date = df.index[-1]
last_actual_value = df['Total_Deposits'].iloc[-1]  # Assuming ~380.03 Billion Lei

# Cumulative sum of forecasted changes, starting from the last actual value
forecast_df['Forecasted Deposits (Billion DKK)'] = forecast_df['Forecasted Deposits (Billion DKK)'].cumsum() + last_actual_value

# Add December 2024 (last actual value) to the forecast table
dec_2024_row = pd.DataFrame({
    'Date': [last_actual_date],
    'Forecasted Deposits (Billion DKK)': [last_actual_value],
    'Conditional Volatility': [None]  # No volatility for the actual value
})

# Combine December 2024 with forecast data
forecast_df = pd.concat([dec_2024_row, forecast_df], ignore_index=True)

# Display the updated forecast table
forecast_df
```

```python
# Confidence intervals (using conditional volatility)
upper_bound = forecast_df['Forecasted Deposits (Billion DKK)'] + 1.96 * forecast_df['Conditional Volatility'].fillna(0)
lower_bound = forecast_df['Forecasted Deposits (Billion DKK)'] - 1.96 * forecast_df['Conditional Volatility'].fillna(0)

# Plotting
plt.figure(figsize=(14, 6))

# Plot actual deposits
plt.plot(df['Total_Deposits'], label='Actual', color='blue')

# Plot forecasted deposits (starting from Dec 2024)
plt.plot(forecast_df['Date'], forecast_df['Forecasted Deposits (Billion DKK)'], label='Forecast', color='orangered', linestyle=':')

# Fill confidence intervals
plt.fill_between(forecast_df['Date'], lower_bound, upper_bound, color='peru', alpha=0.3, label='95% Confidence Interval')

# Adjust the position for Dec 2024 annotation
dec_2024 = forecast_df['Forecasted Deposits (Billion DKK)'].iloc[0]  # Use .iloc to access the value from the first row
plt.annotate(f'{dec_2024:.2f}',
             xy=(pd.to_datetime('2024-12-01'), dec_2024),
             xytext=(pd.to_datetime('2025-01-01'), dec_2024 - 20),
             fontsize=10, color='blue',
             bbox=dict(facecolor='none', edgecolor='none'))

# Adjust the position for Dec 2025 annotation (ARIMA-GARCH)
dec_2025_garch = forecast_df.loc[forecast_df['Date'] == pd.to_datetime('2025-12-01'), 'Forecasted Deposits (Billion DKK)'].values[0]
plt.annotate(f'{dec_2025_garch:.2f}',
             xy=(pd.to_datetime('2025-12-01'), dec_2025_garch),
             xytext=(pd.to_datetime('2025-12-01'), dec_2025_garch + 5),
             fontsize=10, color='orangered',
             bbox=dict(facecolor='none', edgecolor='none'))

# Titles and labels
plt.title('Household Deposits Forecast (ARIMA-GARCH Model)')
plt.xlabel('Date')
plt.ylabel('Deposits (billion DKK)')
plt.legend()
plt.grid(True)
plt.show()
```
